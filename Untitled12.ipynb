```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("personality_dataset.csv")
dataset.head(4)
```

```python
dataset.isnull().sum()
```

```python
dataset["Post_frequency"].fillna(dataset["Post_frequency"].mode()[0],inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit(dataset["Personality"])
dataset["Personality"]=le.transform(dataset["Personality"])
dataset
```

```python
sns.pairplot(data=dataset,hue="Personality")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Personality"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
ss.fit(x)
```

```python
X=pd.DataFrame(ss.transform(x),columns=x.columns)
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=44)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
dt.score(x_train,y_train)*100,dt.score(x_test,y_test)*100
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
svc.fit(x_train,y_train)
```

```python
svc.score(x_train,y_train)*100,svc.score(x_test,y_test)*100
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)*100,lr.score(x_test,y_test)*100
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
vc = VotingClassifier(estimators=[('lr1',lr), ('dt1',dt), ('svc1', svc)],
 voting='hard')
```

```python
vc.fit(x_train,y_train)
```

```python
vc.score(x_train,y_train)*100,vc.score(x_test,y_test)*100
```

```python
prd = {"dt":dt.predict(x_test),"svc":svc.predict(x_test),"lr":lr.predict(x_test)}
```

```python
pd.DataFrame(prd)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
bg = BaggingClassifier(estimator=DecisionTreeClassifier(),n_estimators=100)
```

```python
bg.fit(x_train,y_train)
```

```python
bg.score(x_train,y_train)*100,bg.score(x_test,y_test)*100
```

```python
rf = RandomForestClassifier(n_estimators=100)
```

```python
rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train)*100,rf.score(x_test,y_test)*100
```

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler()
x_resampled, y_resampled = rus.fit_resample(x_train, y_train)
x_resampled, y_resampled
```
